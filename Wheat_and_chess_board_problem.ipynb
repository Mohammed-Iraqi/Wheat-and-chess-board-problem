```python
# [Problem 1] Number of wheat on a 2 x 2 square chess board

"""
Represent a 2 x 2 square chessboard by ndarray and calculate the number of wheat in each square. 
After placing one wheat on the upper left square, proceed to two to the right, four to the lower left,
and eight to the lower right.

"""
# Sample 
import numpy as np
n_squares = 4
small_board_list = [1]
for i in range(n_squares - 1):
    small_board_list.append(2*small_board_list[-1])
print("Arrange wheat on a plate of 4 squares (list)：{}".format(small_board_list))

small_board_ndarray = np.array(small_board_list)
print("Arrange wheat on a plate of 4 squares (ndarray)：{}".format(small_board_ndarray))

# Code of Problem #1
chess_board = small_board_ndarray.reshape((2, 2))
print(chess_board)


# [Problem 2] Expansion to n × m mass
N = 8
M = 8
n_squares = N*M
small_board_list = [1]
for i in range(n_squares - 1):
    small_board_list.append(2*small_board_list[-1])
    
# print("Arrange wheat on a plate of 8 squares (list)：{}".format(small_board_list))

small_board_ndarray = np.array(small_board_list)
# print("Arrange wheat on a plate of 8 squares (ndarray)：{}".format(small_board_ndarray))

# Code of Problem #2
chess_board = small_board_ndarray.reshape((N, M))
print(chess_board)


# [Problem 3] Total number of wheat
print("Sum of all wheat on the chess board：{}".format(chess_board.sum()))

col_sum = []
for i in range(M):
    col_sum.append(chess_board[i,:].sum())
    print(col_sum[i])
    
import matplotlib.pyplot as plt
plt.xlabel("column")
plt.ylabel("number of wheat")
plt.title("number in each column")
plt.bar(range(M),col_sum[:])
plt.show()



#  [Problem 4] Heat map of the number of wheat 
plt.xlabel("column")
plt.ylabel("row")
plt.title("heatmap")
plt.pcolor(small_board_ndarray.reshape(N, M))
plt.show()


# [Problem 5] How many times is the second half as long as the first half?
# First Half
col_sum1 = []
sum1 = 0;
for i in range(0,4,1):
    col_sum1.append(chess_board[i,:].sum())
    sum1 = sum1 + col_sum1[i]

# Second Half
col_sum2 = []
sum2 = 0;
for i in range(4,8,1):
    col_sum1.append(chess_board[i,:].sum())
    sum2 = sum2 + col_sum1[i]
    
print("How many times is the second half as long as the first half: ",sum2/sum1)

# [Problem 6] Creat a function to solve problem #2 of this assingnment?

def Chess_Board_Function(N,M):
    n_squares = N*M
    small_board_list = [1]
    for i in range(n_squares - 1):
        small_board_list.append(2*small_board_list[-1])
        small_board_ndarray = np.array(small_board_list)
    answer = small_board_ndarray.reshape((N, M))
    return answer

# print("Arrange wheat on a plate of 8 squares (ndarray)：{}".format(small_board_ndarray))
N = 8
M = 8
print(Chess_Board_Function(N,M))

#[Problem 7] Comparing calculation times 
import timeit
starttime = timeit.default_timer()
print("The start time is :",starttime)
N = 8
M = 8
print(Chess_Board_Function(N,M))
print("The time difference is :", timeit.default_timer() - starttime)

```
